```python
import numpy as np

# 1. CREATE ARRAYS
arr1D = np.array([10, 20, 30, 40, 50])
arr2D = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("1D Array:")
print(arr1D)
print("\n2D Array:")
print(arr2D)

# 2. OPERASI MATEMATIK
print("\nTambah 5 pada setiap elemen 1D array:")
print(arr1D + 5)

print("\nDarab setiap elemen dalam 2D array dengan 2:")
print(arr2D * 2)

# 3. SHAPE & SIZE
print("\nShape arr2D:", arr2D.shape)
print("Total elements arr2D:", arr2D.size)

# 4. INDEXING
print("\nElemen kedua dalam 1D array:", arr1D[1])
print("Elemen baris 1, kolum 2 dalam 2D array:", arr2D[0, 1])

# 5. SLICE
print("\nSlice arr1D [1:4]:", arr1D[1:4])

# 6. RANDOM ARRAY
rand_arr = np.random.randint(1, 10, (3, 3))
print("\nRandom 3x3 array:")
print(rand_arr)

# 7. SUM, MAX, MIN
print("\nJumlah keseluruhan rand_arr:", rand_arr.sum())
print("Nilai maksimum:", rand_arr.max())
print("Nilai minimum:", rand_arr.min())
```
